 Data Cleaning & Preprocessing

handling missing values (replace NaN with 'unknown')

```python
cols_to_fill = ['director', 'cast', 'country', 'date_added', 'rating', 'duration']
for col in cols_to_fill:
    df[col] = df[col].fillna('Unknown')

print("Missing values handled")
```

```python
print(f"director missing: {df['director'].isna().sum()}") 
```

convert date_added to datetime

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month

print("Date conversion complete")
```

Duration of the Movies and TV shows

```python
def split_duration(duration):
    if pd.isna(duration):  
        return None, None
    duration = str(duration)  
    if 'min' in duration:
        return int(duration.replace(' min', '').replace(' ', '')), None
    elif 'Season' in duration or 'Seasons' in duration:
        seasons = duration.replace(' Season', '').replace(' Seasons', '').replace('s', '').strip()
        return None, int(seasons)
    return None, None
```

```python
duration_split = df['duration'].apply(split_duration)
df['duration_minutes'] = duration_split.apply(lambda x: x[0] if x[0] is not None else np.nan)
df['duration_seasons'] = duration_split.apply(lambda x: x[1] if x[1] is not None else np.nan)
df.drop('duration', axis=1, inplace=True)
```

split genres

```python
df['listed_in'] = df['listed_in'].str.split(', ')
```

## Genre Based Recommendation Engine

```python

def recommend_similar_titles(target_title, df, top_n=5):
    
    
    target_row = df[df['title'].str.contains(target_title, case=False, na=False)]
    if target_row.empty:
        return f"Title '{target_title}' not found in catalog"
    
    target_genres = set(target_row.iloc[0]['listed_in'])
    print(f"Target: '{target_title}'")
    print(f"Genres: {', '.join(target_genres)}")
    
    recommendations = []
    for idx, row in df.iterrows():
        if row['title'] == target_title:
            continue
            
        title_genres = set(row['listed_in']) if row['listed_in'] else set()
        overlap_count = len(target_genres.intersection(title_genres))
        
        if overlap_count > 0:
            recommendations.append({
                'title': row['title'],
                'type': row['type'],
                'overlap_score': overlap_count,
                'genres': row['listed_in'],
                'year': row['release_year']
            })
    
    recommendations.sort(key=lambda x: (-x['overlap_score'], -x['year']))
    return recommendations[:top_n]


print("="*60)
print(" GENRE-BASED RECOMMENDATION ENGINE")
print("="*60)

# Test 1: Popular drama
recs1 = recommend_similar_titles("House of Cards", df)
print("\n Drama recommendations for 'House of Cards':")
for i, rec in enumerate(recs1, 1):
    print(f"   {i}. {rec['title']} ({rec['type']}) - {rec['overlap_score']} genres match")

# Test 2: Comedy
recs2 = recommend_similar_titles("The Office", df)
print("\n Comedy recommendations for 'The Office':")
for i, rec in enumerate(recs2, 1):
    print(f"   {i}. {rec['title']} ({rec['type']}) - {rec['overlap_score']} genres match")

# Test 3: User input demo 
print("\n TRY YOUR OWN:")
user_title = input("Enter any Netflix title: ")
user_recs = recommend_similar_titles(user_title, df)
for i, rec in enumerate(user_recs, 1):
    print(f"   {i}. {rec['title']} ({rec['type']})")

print(" Built content recommendation using Jaccard similarity on genre overlap")
```
